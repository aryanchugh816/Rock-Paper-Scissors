```python
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
import cv2
```

```python
train_gen = ImageDataGenerator(rescale=1/255.0, validation_split=0.1, zoom_range=0.4)
```

```python
train_generator = train_gen.flow_from_directory("training_data_2/", batch_size=128, subset='training')
```

```python
train_generator.image_shape
```

```python
for (x,y) in train_generator:
    for i in range(10):
        print(y[i])
        plt.imshow(x[i])
        plt.show()
        
    break
```

```python
val_generator = train_gen.flow_from_directory("Training_data_2/", subset='validation')
```

```python
# Model
from keras.models import Sequential
from keras.layers import *
```

```python
model = Sequential()
```

```python
model.add(Conv2D(32, (3,3), activation='relu', input_shape=(256,256,3)))
model.add(MaxPool2D((2,2)))
model.add(Conv2D(32, (3,3), activation='relu'))
model.add(MaxPool2D((2,2)))
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPool2D((2,2)))
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPool2D((2,2)))
model.add(Conv2D(128, (3,3), activation='relu'))
model.add(MaxPool2D((2,2)))
model.add(Conv2D(128, (3,3), activation='relu'))
model.add(MaxPool2D((2,2)))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(3, activation='softmax'))
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
```

```python
hist = model.fit_generator(
    train_generator,
    steps_per_epoch=5,
    epochs=15,
    validation_data=val_generator,
    validation_steps=2
)
```

```python
model.metrics_names
```

```python
test_gen = ImageDataGenerator(rescale=1/255.0)
test_generator = test_gen.flow_from_directory("Training_data_2/")
test_generator.class_indices
```

```python
model.evaluate_generator(test_generator)
```
